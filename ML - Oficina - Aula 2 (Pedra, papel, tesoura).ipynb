```python
import random
jogar_de_novo = True
opcoes = [1,2,3]
escolhas = []

def jogo():
    while True:
        user_jogada = int(input("Digite:\n1 para Pedra\n2 para Papel\n3 para Tesoura\n4 para Sair\n"))
        computador = random.randint(1,3)
        a = escolhas.count(1)
        b = escolhas.count(2)
        c = escolhas.count(3)
        pesos = [a, b, c]
        print("\nVocê escolheu", user_jogada, "e o Computador escolheu", computador)
        if len(escolhas) < 4:
            computador = random.randint(1,3)
            print("\nVocê escolheu", user_jogada, "e o Computador escolheu", computador)
            if user_jogada == computador:
                print("Empate!")
                escolhas.append(user_jogada)
            elif user_jogada == 1:
                if computador == 2:
                    print("Papel cobre pedra! Você perdeu!")
                else:
                    print("Pedra quebra tesoura! Você ganhou!")
                escolhas.append(user_jogada)
            elif user_jogada == 2:
                if computador == 1:
                    print("Papel cobre pedra! Você ganhou!")
                else:
                    print("Tesoura corta papel! Você perdeu!")
                escolhas.append(user_jogada)
            elif user_jogada == 3:
                if computador == 1:
                    print("Pedra quebra tesoura! Você perdeu!")
                else:
                    print("Tesoura corta papel! Você ganhou!")
                escolhas.append(user_jogada)
            else:
                break
        else:
            computador = random.choices(opcoes, weights = pesos)
            escolhas.append(user_jogada)
```
